# tibble의 생성
### 기존의 데이터 프레임을 tibble로 변환
+ 함수 as_tibble()

```r
library(tidyverse)
as_tibble(cars)
```

### 개별 벡터를 이용한 tibble의 생성
+ 길이가 1인 스칼라만 순환법칙 적용
+ 함께 입력되는 변수를 이용한 다른 변수의 생성 가능
+ 열(변수) 단위로 입력

```r
tibble(x=1:3,y=x+1,z=1)
```

### tibble() vs data.frame()
+ 함께 입력된 변수를 이용하여 다른 변수를 만드는 기능

```r
data.frame(x:1:3,y=x+1) # 오류
```

+ 문자형 벡터를 요인으로 전환하지 않음

```r
tibble(x=1:3, y=letters[1:3])
```

```r
str(data.frame(x=1:3,y=letters[1:3]))
```

### 행 단위로 입력하여 tibble 생성
+ 첫 줄 : 변수 이름은 '~'로 시작
+ 각 자료는 콤마로 구분

```r
tribble(
~x,~y,
1, "a",
2, "b",
3, "c")
```

## tibble과 전통적 데이터 프레임의 비교
### 출력 방식의 차이
+ 전통적 데이터 프레임 : 가능한 모든 자료를 홤현에 출력. 대규모 자료의 경우 내용 확인에 어려움

```r
data(Cars93, package="MASS")
Cars93
```

+ tibble을 사용하면 처음 10개 데이터만 출력

```r
as_tibble(Cars93)
```

+ 변수 이름과 더불어 변수의 유형을 함께 표시
+ 더 많은 자료 확인
+ print(tbl, n=20, width=Inf)

```r
print(as_tibble(Cars93), n=3, width=Inf)
```

### Row names 처리 방식의 차이
+ 전통적 데이터 프레임 : 자료 출력 시 row name 함께 출력
+ Tibble : 생략

```r
head(mtcars)
```

```r
mtcasr_t <- as_tibble(mtcars)
```

```r
print(mtcars_t, n=6)
```

+ 생략된 row name : 변수로 전환

```r
mtcars_d <- rownames_to_column(mtcars, var="rowname")
mtcars_t <- as_tibble(mtcars_d)
mtcars_t
```

### 인덱싱 방법의 차이
+ 달라 기호를 이용하는 경우 : 변수 이름의 부분 매칭 허용 여부

```r
df1 <- data.frame(xyz=1:3, abc=letters[1:3])
df1$x
```

```r
tb1 <- as_tibble(df1)
tb1$x
tb1$xyz
```

+ 행렬 방식의 인덱싱 결과
+ 전통적인 데이터 프레임 : 선택되는 변수의 개수에 따라 벡터 혹은 데이터 프레임
+ tibble : 선택되는 변수의 개수와 관계 없이 항상 tibble 유지

```r
mtcars[, 1:2]
mtcars[, 1]

mtcasr_t[, 1:2]
mtcasr_t[, 1]
mtcasr_t[1, 1]
```
